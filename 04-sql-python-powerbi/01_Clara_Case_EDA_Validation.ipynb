# Clara Case Study – Exploratory Data Analysis

## Objective
The CEO is concerned about low sales volume, especially from new users.  
The objective of this analysis is to:

Validate data integrity and understand the structure of the conversion funnel before building the production model.

Data corresponds to one day of operations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

### LOAD DATA

```python
home = pd.read_csv("data/raw/home_page_table.csv")
search = pd.read_csv("data/raw/search_page_table.csv")
payment = pd.read_csv("data/raw/payment_page_table.csv")
confirmation = pd.read_csv("data/raw/payment_confirmation_table.csv")
users = pd.read_csv("data/raw/user_table.csv")

print("Home:", home.shape)
print("Search:", search.shape)
print("Payment:", payment.shape)
print("Confirmation:", confirmation.shape)
print("Users:", users.shape)
```

### STRUCTURE OVERVIEW

```python
datasets = {
    "home": home,
    "search": search,
    "payment": payment,
    "confirmation": confirmation,
    "users": users
}

for name, df in datasets.items():
    print(f"\n{name.upper()}")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    display(df.head())
```

### DATA QUALITY CHECK

```python
for name, df in datasets.items():
    print(f"\n{name.upper()} DATA QUALITY")
    print("Null values:\n", df.isnull().sum())
    print("Duplicate rows:", df.duplicated().sum())
    print("Unique user_ids:", df['user_id'].nunique())
```

### FUNNEL CONSISTENCY VALIDATION

```python
home_users = set(home['user_id'])
search_users = set(search['user_id'])
payment_users = set(payment['user_id'])
confirmation_users = set(confirmation['user_id'])

print("Search users not in Home:", len(search_users - home_users))
print("Payment users not in Search:", len(payment_users - search_users))
print("Confirmation users not in Payment:", len(confirmation_users - payment_users))
```

### FUNNEL VOLUME

```python
home_count = len(home_users)
search_count = len(search_users)
payment_count = len(payment_users)
confirmation_count = len(confirmation_users)

print("Home:", home_count)
print("Search:", search_count)
print("Payment:", payment_count)
print("Confirmation:", confirmation_count)
```

### PRELIMINARY CONVERSION RATES

```python
print("Home → Search:", round(search_count / home_count, 4))
print("Search → Payment:", round(payment_count / search_count, 4))
print("Payment → Confirmation:", round(confirmation_count / payment_count, 4))
print("Overall Conversion Rate:", round(confirmation_count / home_count, 4))
```

### Registration Date Distribution

Since the user table contains a range of registration dates,
we validate whether the dataset includes historical users.

```python
print("Min registration date:", users['date'].min())
print("Max registration date:", users['date'].max())
print("Unique registration dates:", users['date'].nunique())
```
